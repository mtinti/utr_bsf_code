```python
db = gffutils.FeatureDB(dbfn='bw_file/TbruceiTREU927_v50.db',keep_order=True)

#these are the track for the barcoded reads
#in the blasticidin (positive) experimenr
barcoded_g = [
    'bw_file/all_BSF_F_FR_G.bw',
    'bw_file/all_BSF_R_RR_G.bw',
]

#these are the track for the barcoded reads
#in the ganciclovir (negative) experimenr
barcoded_b = [
    'bw_file/all_BSF_R_RR_B.bw',
    'bw_file/all_BSF_F_FR_B.bw']

#we use this function to compute the max value of the 
#bigwig coverage file in a window. We want the barcoded 
#coverage values to be o the same scale for 
#the positive and negative experiments
def get_bigwig_max(
    #the chromosome of the reagion of interest
    chro="Tb927_06_v5.1", 
    #the start of the reagion of interest
    start=425312,
    #the end of the reagion of interest
    end=429872,
    #list of grouped file paths
    infiles=[]):
    #store results
    values = []
    
    #retrieve the max coverage value in each track
    for f in infiles:
        bw = pyBigWig.open(f)
        values.append(bw.stats(chro, start, end, type="max")[0])
        #print(f,bw.stats(chro, start, end, type="max")[0])
    max_values = int(max(values))
    
    #incrase the max value
    #of 10% to make a nicer plot
    #that won't cut the upper part of the signal
    max_values = max_values + (max_values*10/100)
    return str(int(max_values))

#we get a string with a region of interest to use
#with pyGenomeTracks. we want to query a gene of interest and
#get a region that cover one gene before and one gene after the
#gene of interest
def get_reference_string(gene_id='Tb927.11.11770'):
    '''
    given a gene id this function will create
    a string containing the genomic reagion to plot
    '''
    
    #grab the gene from the database
    gene = db[gene_id]
    #print(gene.id, gene.start, gene.end, gene.seqid)
    
    #grab from the database the first 3 gene upstream
    #and downstrem the gene of interest and put
    #info in a dataframe
    
    #the sql query for the database
    template = """SELECT features.id,features.start,features.end,features.seqid,features.strand FROM features WHERE
    features.{position} {direction} {input_base} AND
    features.seqid == '{seqid}' AND 
    features.featuretype == 'CDS' 
    ORDER BY features.start {order}
    LIMIT 10
    """

    cols = ['geneId','geneStart','geneEnd','seqId','strand']
    query_string = template.format(
        #the site of the query features
        input_base=gene.end,#1527246, 
        #start or end of the reference genes
        position = 'start',
        #chromosome of the query genes
        seqid=gene.seqid,#'Tb927_10_v5.1',
        #select the comparison between the
        #input_base and the reference genes
        direction = '>', order ='ASC')

    results_down= pd.DataFrame.from_records(data = db.execute(query_string).fetchall(),
                                           columns=cols)

    #print(results_down)
    query_string = template.format(
        #the site of the query features
        input_base=gene.start,#1521347, 
        #start or end of the reference genes
        position = 'end',
        #chromosome of the query genes
        seqid=gene.seqid,#'Tb927_10_v5.1',
        #select the comparison between the
        #input_base and the reference genes
        direction = '<', order ='DESC')

    results_up = pd.DataFrame.from_records(data = db.execute(query_string).fetchall(),
                                          columns=cols)


     
    peak_name = gene.id
    chro = gene.seqid
    #to have one gene at the front (+1) and at the back (-2) of the
    #gene of interest we grab the end and start of the
    #genes at -2 and at +2
    if results_up.shape[0] == 0:
        # we are at the end of the chromosome
        start = gene.start
    else:   
        results_up['geneId']=[n.split(':')[0] for n in results_up['geneId']]
        results_up=results_up[ (results_up['geneId'].str.startswith('Tb927.') | results_up['geneId'].str.startswith('Tb11.') )]
        results_up=results_up.drop_duplicates(subset='geneId')
        results_up.reset_index(inplace=True,drop=True)        
        if results_up.shape[0] == 1:
            start = results_up.iloc[0]['geneEnd']        
        else:       
            start = results_up.iloc[1]['geneEnd']
        
    if results_down.shape[0] == 0:
        # we are at the end of the chromosome
        end = gene.end 
    else:
        results_down['geneId']=[n.split(':')[0] for n in results_down['geneId']]
        results_down=results_down[ ( results_down['geneId'].str.startswith('Tb927.') | results_down['geneId'].str.startswith('Tb11.'))]
        results_down=results_down.drop_duplicates(subset='geneId')
        results_down.reset_index(inplace=True,drop=True)   
        if results_down.shape[0] == 1:
            end = results_down.iloc[0]['geneStart']        
        else:       
            end = results_down.iloc[1]['geneStart']        
        

    rep_string = '{chro}:{start}-{end}'.format(chro=chro,start=start,end=end)

    rep_string, peak_name
    return chro,start,end,rep_string
    
```

```python
#find all the protein coding genes on main chromosomes
#keep track of orientation
main_chro =  [
 'Tb927_01_v5.1',
 'Tb927_02_v5.1',
 'Tb927_03_v5.1',
 'Tb927_04_v5.1',
 'Tb927_05_v5.1',
 'Tb927_06_v5.1',
 'Tb927_07_v5.1',
 'Tb927_08_v5.1',
 'Tb927_09_v5.1',
 'Tb927_10_v5.1',
 'Tb927_11_v5.1']

def parse_gff(gff_file):
    genes_with_cds = []
    current_gene = None
    current_gene_has_cds = False

    with open(gff_file, 'r') as file:
        for line in file:
            if line.startswith('#'):
                continue

            columns = line.strip().split('\t')
            chromosome = columns[0]
            feature_type = columns[2]
            start = int(columns[3])
            end = int(columns[4])
            strand = columns[6]
            attributes = columns[8].split(';')

            attribute_dict = {}
            for attribute in attributes:
                key, value = attribute.split('=')
                attribute_dict[key] = value

            if feature_type == 'gene':
                if current_gene is not None and current_gene_has_cds:
                    gene_id = current_gene['attributes']['ID']
                    genes_with_cds.append((gene_id, current_gene['strand'], current_gene['chromosome']))

                current_gene = {
                    'start': start,
                    'end': end,
                    'strand': strand,
                    'chromosome': chromosome,
                    'attributes': attribute_dict,
                }
                current_gene_has_cds = False
            elif current_gene is not None and current_gene['start'] <= start and current_gene['end'] >= end:
                if feature_type == 'CDS':
                    current_gene_has_cds = True
            else:
                if current_gene is not None and current_gene_has_cds:
                    gene_id = current_gene['attributes']['ID']
                    genes_with_cds.append((gene_id, current_gene['strand'], current_gene['chromosome']))

                current_gene = None
                current_gene_has_cds = False

        if current_gene is not None and current_gene_has_cds:
            gene_id = current_gene['attributes']['ID']
            genes_with_cds.append((gene_id, current_gene['strand'], current_gene['chromosome']))

    return genes_with_cds



gff_file = 'bw_file/merged_fixed.gff'
genes_with_cds = parse_gff(gff_file)
len(genes_with_cds)
```

```python
#this function take in inputs a pyGenomeTracks tempate file
#and modify it with the max coverage value of the barcodes
rep_string_dictionary = {}
def create_templates(gene_id='Tb927.11.11770', 
                     template='',
                     out_files = '',
                     prefix=''):
    chro,start,end,rep_string = get_reference_string(gene_id=gene_id)

    max_barcoded_b = get_bigwig_max(chro,start,end, barcoded_b)
    max_barcoded_g = get_bigwig_max(chro,start,end, barcoded_g)

    template = open(template).read()
    template = template.replace('{max_bla}',max_barcoded_b)
    template = template.replace('{max_gla}',max_barcoded_g)
    open(out_files+prefix+'{gene_id}.ini'.format(gene_id=gene_id),'w').write(template)
    rep_string_dictionary[gene_id]=rep_string

#now we create templates for 2 genes with CDS that will plots all the avaible tracks
#when a gene is in the 5' to 3' orientation we use the template_all_F
#when a gene is in the 3' to 5' orientation we use the template_all_R
#template_all_R flips the coverage tracks
orient_dictionary = {}
for n in genes_with_cds:
    orient_dictionary[n[0]] = n[1]

gene_of_interest = [('Tb927.1.710', '-', 'Tb927_01_v5.1'),
               ('Tb927.10.5620', '+', 'Tb927_10_v5.1')]

for n in tqdm(gene_of_interest): # replace the list with genes_with_cds to make figures for all genes
    count = 0
    if n[-1] in main_chro:
        count+=1
        orient = orient_dictionary[n[0]]
        if orient == '+':
            template_file = 'bw_file/template_all_F.ini'

        else:
            template_file = 'bw_file/template_all_R.ini'

        create_templates(gene_id=n[0], 
                         template=template_file,
                         out_files='bw_file/full_templates/')
```

```python
# we used this function to plot the coverage tracks for all genes
# here we will show this for the two of the selected genes above
from concurrent.futures import ProcessPoolExecutor
from tqdm.auto import tqdm
from IPython.display import Image

# Function to process each peak_name
def process_peak(peak_name):
    rep_string = get_reference_string(peak_name)[-1]
    orient = orient_dictionary[peak_name]
    file_name = f'bw_file/full_images/{peak_name}_coverage.png'

    if orient == '+':
        !pyGenomeTracks --tracks "bw_file/full_templates/"$peak_name".ini" \
        --region $rep_string --fontSize 6 --title $rep_string \
        --dpi 300 --plotWidth 24 --trackLabelFraction 0 \
        -o $file_name 2> /dev/null
    else:
        !pyGenomeTracks --tracks "bw_file/full_templates/"$peak_name".ini" \
        --region $rep_string --fontSize 6 --title $rep_string \
        --dpi 300 --plotWidth 24 --trackLabelFraction 0 --decreasingXAxis \
        -o $file_name 2> /dev/null

    return file_name

process_peak('Tb927.10.5620')
```

```python
# show versions of packages
# adopted from https://stackoverflow.com/questions/40428931/package-for-listing-version-of-packages-used-in-a-jupyter-notebook
import pkg_resources
import types
import sys
def get_imports():
    for name, val in globals().items():
        if isinstance(val, types.ModuleType):
            # Split ensures you get root package, 
            # not just imported function
            name = val.__name__.split(".")[0]
        elif isinstance(val, type):
            name = val.__module__.split(".")[0]
        # Some packages are weird and have different
        # imported names vs. system/pip names. Unfortunately,
        # there is no systematic way to get pip names from
        # a package's imported name. You'll have to add
        # exceptions to this list manually!
        poorly_named_packages = {
            "sklearn": "scikit-learn"
        }
        if name in poorly_named_packages.keys():
            name = poorly_named_packages[name]
        yield name.lower()
imports = list(set(get_imports()))

# The only way I found to get the version of the root package
# from only the name of the package is to cross-check the names 
# of installed packages vs. imported packages
modules = []
for m in sys.builtin_module_names:
    if m.lower() in imports and m !='builtins':
        modules.append((m,'Python BuiltIn'))
        imports.remove(m.lower())

for m in pkg_resources.working_set:
    if m.project_name.lower() in imports and m.project_name!="pip":
        modules.append((m.project_name, m.version))
        imports.remove(m.project_name.lower())

for m in sys.modules:
    if m.lower() in imports and m !='builtins':
        modules.append((m,'unknown'))

# print('System=='+platform.system()+' '+platform.release()+'; Version=='+platform.version())
for r in modules:
    if 'unknown' in r[1]:
        continue
    print("{}=={}".format(*r))
```
