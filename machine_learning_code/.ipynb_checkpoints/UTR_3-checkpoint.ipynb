```python
import re
class ExtractDNAFeatures:
    def __init__(self, dna_sequence, dna_sequence_id):
        self.dna_sequence = dna_sequence
        self.dna_sequence_id = dna_sequence_id
        self.sequence_length = len(dna_sequence)
        

    def base_count(self):
        
        base_counts = {'A': 0, 'T': 0, 'C': 0, 'G': 0}

        for base in self.dna_sequence:
            if base in base_counts:
                base_counts[base] += 1

        base_counts['AT'] = base_counts['A'] + base_counts['T']
        base_counts['CG'] = base_counts['C'] + base_counts['G']
        base_counts['CT'] = base_counts['C'] + base_counts['T']
        base_counts['AG'] = base_counts['A'] + base_counts['G'] 
        
        return base_counts

    def count_tracts(self, pattern, min_length=5):
        regex_pattern = '[' + pattern + ']{' + str(min_length) + ',}'
        tracts = re.findall(regex_pattern, self.dna_sequence)
        return len(tracts)

    def longest_tract_length(self, pattern, min_length=5, max_mismatch=0):
        regex_pattern = '[' + pattern + ']{' + str(min_length) + ',}'
        tracts = list(re.finditer(regex_pattern, self.dna_sequence))
        if len(tracts) == 0:
            return 0, None, 0

        tracts_indexes = [(x.start(), x.end()) for x in tracts]
        new_tracts_indexes = []
        current_tract = tracts_indexes[0]

        for i in range(1, len(tracts_indexes)):
            if tracts_indexes[i][0] - current_tract[1] <= max_mismatch:
                current_tract = (current_tract[0], tracts_indexes[i][1])
            else:
                new_tracts_indexes.append(current_tract)
                current_tract = tracts_indexes[i]

        new_tracts_indexes.append(current_tract)
        longest_tract = max(new_tracts_indexes, key=lambda x: x[1] - x[0])
        longest_length = longest_tract[1] - longest_tract[0]
        start_index = longest_tract[0]
        
        return longest_length, start_index, len(new_tracts_indexes)
    
    def count_motif(self, motif='AAGAAAA'):
        return len(re.findall(motif, self.dna_sequence))
                   
    
    def get_features(self):
        base_counts = self.base_count()
        features = {'count_' + key: value for key, value in base_counts.items()}
        
        
        #polypyrimidine_pattern = '[CT]{' + str(min_length) + ',}'
        #polypurine_pattern = '[AG]{' + str(min_length) + ',}'
        #for base in ['A','T','G','C', 'CT', 'AG', 'AT', 'GC']:
        #    count_tracts = self.count_tracts(base)
        #    features[f'count_tracts_{base}'] = count_tracts

        #for base in ['A','T','G','C', 'CT', 'AG', 'AT', 'GC']:
        #    count_tracts = self.longest_tract_length(base, max_mismatch=0)
        #    features[f'longest_tract_{base}'] = count_tracts[0]
        
        '''
        for base in ['A', 'T', 'G', 'C', 'CT', 'AG']:
            max_mismatch=4
            count_tracts = self.longest_tract_length(base, max_mismatch=max_mismatch)
            #print(count_tracts)
            features[f'len_longest_tract_{base}_mismatch_{max_mismatch}'] = count_tracts[0]
            features[f'count_tracts_{base}_mismatch_{max_mismatch}'] = count_tracts[2]

        
        for base in ['A', 'T', 'G', 'C', 'CT', 'AG']:
            max_mismatch=3
            count_tracts = self.longest_tract_length(base, max_mismatch=max_mismatch)
            #print(count_tracts)
            features[f'len_longest_tract_{base}_mismatch_{max_mismatch}'] = count_tracts[0]
            features[f'count_tracts_{base}_mismatch_{max_mismatch}'] = count_tracts[2]

        '''
        for base in ['A',  'C', 'CT', 'AG']:
            max_mismatch=2
            count_tracts = self.longest_tract_length(base, max_mismatch=max_mismatch)
            #print(count_tracts)
            #features[f'len_longest_tract_{base}_mismatch_{max_mismatch}'] = count_tracts[0]
            features[f'count_tracts_{base}_mismatch_{max_mismatch}'] = count_tracts[2]

        for base in ['A', 'C', 'CT', 'AG']:
            max_mismatch=1
            count_tracts = self.longest_tract_length(base, max_mismatch=max_mismatch)
            #features[f'len_longest_tract_{base}_mismatch_{max_mismatch}'] = count_tracts[0]
            features[f'count_tracts_{base}_mismatch_{max_mismatch}'] = count_tracts[2]  
            
        for base in ['A', 'C', 'CT', 'AG']:
            max_mismatch=0
            count_tracts = self.longest_tract_length(base, max_mismatch=max_mismatch)
            #features[f'len_longest_tract_{base}_mismatch_{max_mismatch}'] = count_tracts[0]
            features[f'count_tracts_{base}_mismatch_{max_mismatch}'] = count_tracts[2] 
            
        
            
            
            
        # Update the values by dividing them by the DNA sequence length
        for key in features:
            features[key] /= self.sequence_length
            
        #features['count_motif_AAGAAAA']=self.count_motif('AAGAAAA')
        
        features['seq_len']=self.sequence_length
        
        return {self.dna_sequence_id:features}
        
```

```python
import numpy as np
import pandas as pd

# Read in your data
#y_df = pd.read_csv('translation_efficency/12864_2014_6600_MOESM3_ESM.txt', sep='\t')

# Function to perform Monte Carlo simulation for error propagation in a ratio
def simulate_ratio_error(r_mean, r_std, m_mean, m_std, num_simulations=10000):
    # Generate simulated data assuming normal distribution of r and m
    simulated_r = np.random.normal(loc=r_mean, scale=r_std, size=num_simulations)
    simulated_m = np.random.normal(loc=m_mean, scale=m_std, size=num_simulations)

    # Calculate simulated TE values
    simulated_te = simulated_r / (simulated_r + simulated_m)

    # Calculate the standard deviation of simulated TE values
    te_std = simulated_te.std()

    return te_std

# Calculate the means and standard deviations for r and m
r_means = y_df[['cBF1r', 'cBF2r', 'cBF3r']].mean(axis=1)
m_means = y_df[['cBF1m', 'cBF2m', 'cBF3m']].mean(axis=1)
r_stds = y_df[['cBF1r', 'cBF2r', 'cBF3r']].std(axis=1)
m_stds = y_df[['cBF1m', 'cBF2m', 'cBF3m']].std(axis=1)

# Calculate standard deviation via Monte Carlo simulation for each gene
te_std = np.array([simulate_ratio_error(r_mean, r_std, m_mean, m_std) for r_mean, r_std, m_mean, m_std in zip(r_means, r_stds, m_means, m_stds)])

# Calculate standard error of TE
y_df['TE_SE'] = te_std / np.sqrt(3)  # Adjust the divisor to the number of replicates if necessary

# Now y_df['TE_SE'] contains the standard error for the TEperc values
```

```python
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from scipy.stats import pearsonr
from sklearn.metrics import mean_squared_error
def clean(pred, y_test):
    temp=pd.DataFrame()
    temp['pred']=pred
    temp['y_test']=y_test
    temp=temp.replace(-np.inf,np.nan).replace(np.inf,np.nan).dropna()
    return temp['pred'].values,temp['y_test'].values


def abline(pred, y_test, ax, do_clean=True):
    if do_clean:
        
        pred, y_test = clean(pred, y_test)
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(pred, y_test)
    """Plot a line from slope and intercept"""
    #axes = ax.get_gca()
    x_vals = np.array(ax.get_xlim())
    y_vals = intercept + slope * x_vals
    ax.plot(x_vals, y_vals, '--',c='r')
    MSE = mean_squared_error(pred, y_test)
    #r2 = r2_score(pred, y_test)
    r2=r_value**2
    p = pearsonr(pred, y_test)
    
    ax.text(0.01, 0.95, 'MSE: {:.4f}'.format(MSE), fontsize=12, transform=ax.transAxes)
    ax.text(0.01, 0.9, 'r2: {:.4f}'.format(r2), fontsize=12, transform=ax.transAxes)
    #ax.text(0.01, 0.85, 'corr: {:.4f}'.format(r_value), fontsize=12, transform=ax.transAxes)
    ax.text(0.01, 0.85, 'pearson: {:.4f}'.format(p[0]), fontsize=12, transform=ax.transAxes)
    
    #print('mean_squared_error',MSE)
```

```python
# show versions of packages
# adopted from https://stackoverflow.com/questions/40428931/package-for-listing-version-of-packages-used-in-a-jupyter-notebook
import pkg_resources
import types
import sys
def get_imports():
    for name, val in globals().items():
        if isinstance(val, types.ModuleType):
            # Split ensures you get root package, 
            # not just imported function
            name = val.__name__.split(".")[0]
        elif isinstance(val, type):
            name = val.__module__.split(".")[0]
        # Some packages are weird and have different
        # imported names vs. system/pip names. Unfortunately,
        # there is no systematic way to get pip names from
        # a package's imported name. You'll have to add
        # exceptions to this list manually!
        poorly_named_packages = {
            "sklearn": "scikit-learn"
        }
        if name in poorly_named_packages.keys():
            name = poorly_named_packages[name]
        yield name.lower()
imports = list(set(get_imports()))

# The only way I found to get the version of the root package
# from only the name of the package is to cross-check the names 
# of installed packages vs. imported packages
modules = []
for m in sys.builtin_module_names:
    if m.lower() in imports and m !='builtins':
        modules.append((m,'Python BuiltIn'))
        imports.remove(m.lower())

for m in pkg_resources.working_set:
    if m.project_name.lower() in imports and m.project_name!="pip":
        modules.append((m.project_name, m.version))
        imports.remove(m.project_name.lower())

for m in sys.modules:
    if m.lower() in imports and m !='builtins':
        modules.append((m,'unknown'))

# print('System=='+platform.system()+' '+platform.release()+'; Version=='+platform.version())
for r in modules:
    print("{}=={}".format(*r))
```
